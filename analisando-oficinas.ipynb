```python
# Importando Bibliotecas
import pandas as pd
```

```python
df = pd.read_csv('tabela-de-oficinas.csv', encoding='utf-8', sep="\t", header=None)

df.columns = ['Nome', 'Protocolo', 'Mes', 'Resultado']
```

```python
df_sem_nome_e_mes = df.drop(['Nome', 'Mes'], axis=1,inplace=False)
```

```python
df_sem_nome_e_mes
```

```python

df_recomendado = df_sem_nome_e_mes[df_sem_nome_e_mes['Resultado'] == 'RECOMENDADO']
print(df_recomendado)
```

```python
df_recomendado
```

```python
df_nao_recomendado = df_sem_nome_e_mes[df_sem_nome_e_mes['Resultado'] == 'NÃO RECOMENDADO']
print(df_nao_recomendado)
```

```python
df_nao_recomendado
```

```python
def total_de_protocolos(df):
    return df['Protocolo'].count()

def calculo_porcentagem_recomendado(df):
    return len(df_recomendado) / total_de_protocolos(df) * 100

def calculo_porcentagem_nao_recomendado(df):
    return len(df_nao_recomendado) / total_de_protocolos(df) * 100

def nova_tabela(df):
    return total_de_protocolos(df), calculo_porcentagem_recomendado(df), calculo_porcentagem_nao_recomendado(df)

nova_tabela(df)
pd.concat([df_sem_nome_e_mes, df_recomendado, df_nao_recomendado])
```
